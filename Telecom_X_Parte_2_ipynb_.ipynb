Modelo 2: Random Forest



Por que escolher?

Random Forest é um conjunto de árvores de decisão.

Modelos baseados em árvores não são sensíveis à escala das variáveis, então a normalização não é necessária.

É muito eficiente para capturar relações complexas e interações não lineares entre variáveis.

```python
from sklearn.preprocessing import OneHotEncoder
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Remover customerID
X = df.drop(['Churn', 'customerID'], axis=1)
y = df['Churn']

# Converter variáveis categóricas em numéricas (OneHotEncoding)
X = pd.get_dummies(X, drop_first=True)

# Divisão treino/teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Aplicar SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Random Forest
rf = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
rf.fit(X_train_res, y_train_res)

# Predições
y_pred = rf.predict(X_test)

# Avaliação
print("Random Forest com SMOTE:\n")
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print("Matriz de Confusão:\n", cm)

# Representação gráfica
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predito')
plt.ylabel('Real')
plt.title('Matriz de Confusão - Random Forest com SMOTE')
plt.show()
```

Modelo mais complexo, mas sem balanceamento de classes, ele tende a favorecer a classe majoritária.

*  Random Forest com SMOTE:

  Aumentou recall da classe minoritária (identificação de churns), mas precisou sacrificar precisão → trade-off comum.

Indicação de melhor equilíbrio, ainda sem sinais claros de overfitting.

### **Conclusões e ajustes recomendados**

_ Melhor modelo geral: **Random Forest com SMOTE**, pois consegue identificar mais churns (recall maior) sem perder muito em acurácia.

*Ajustes possíveis*:

Para Regressão Logística: aplicar SMOTE ou ponderar classes (class_weight='balanced') para melhorar recall da classe minoritária.

Para Random Forest: ajustar hiperparâmetros (n_estimators, max_depth, min_samples_leaf) e usar validação cruzada para otimizar performance.

Monitorar overfitting: comparar desempenho treino vs teste; caso Random Forest fique muito melhor no treino, reduzir profundidade ou aumentar min_samples_leaf.
